```python
import shutil
# important copy dataset from input to working
# if you don't the laoding of the images will be much slower
source_dir = "/kaggle/input/datasets/bartoszkrolik/whole-sliced-nomad/sliced_data" # paths best to copy from notebook itself
dest_dir = "/kaggle/working/sliced_data"

shutil.copytree(source_dir, dest_dir, dirs_exist_ok=True)
```

```python
import yaml

dataset_path = '/kaggle/working/sliced_data'

data_yaml = {
    'path': dataset_path,
    'train': 'images/train',
    'val': 'images/val',
    'nc': 1, 
    'names': ['person'] 
}

with open('/kaggle/working/data.yaml', 'w') as f:
    yaml.dump(data_yaml, f)
    
print("data.yaml successfully created!")
```

```python
from ultralytics import YOLO

model = YOLO("yolo26n.pt")

results = model.train(
    data="/kaggle/working/data.yaml", 
epochs=70,               
    imgsz=1024,               
    batch=8,                  
    device=[0,1], # IMPORTANT if you use double GPU on kaggle this should be set to [0,1] otherwise [0] it wil use only one GPU
    augment=True,             
    degrees=90.0,             
    flipud=0.5,               
    fliplr=0.5,               
    cache=False,              
    workers=4,                
    optimizer='AdamW',        
    name='nomad_yolo26n-stable'
    
)
```
